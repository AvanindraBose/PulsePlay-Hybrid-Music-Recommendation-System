```python
import kagglehub
import os
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Download latest version
path = kagglehub.dataset_download("undefinenull/million-song-dataset-spotify-lastfm")

print("Path to dataset files:", path)
```

```python
for item in os.listdir(path):
    print(item)
```

```python
data_path = Path(path)
```

```python
songs_data_path = data_path / 'Music Info.csv'
users_data_path = data_path / 'User Listening History.csv'
```

```python
df_songs = pd.read_csv(songs_data_path)
df_songs.head()
```

```python
df_songs.shape
```

```python
df_user = pd.read_csv(users_data_path)
df_user.head()
```

```python
df_user.shape
```

```python
root_path = Path().cwd().parent
song_path_to_save = root_path / "data"/"raw"/"songs-info.csv"
user_path_to_save = root_path / "data"/"raw"/"user-info.csv"
```

```python
df_songs.columns
```

```python
df_songs.rename(columns={'spotify_preview_url':'pulse_play_preview_url', 'spotify_id':'pulse_play_id'},inplace=True)
```

```python
df_songs.to_csv(song_path_to_save,index=False)
df_user.to_csv(user_path_to_save,index=False)
```
